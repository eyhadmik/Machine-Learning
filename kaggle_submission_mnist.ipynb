```python
# pandasÎ•º importÌïòÍ≥† train dataÏôÄ test dataÎ•º Î∂àÎü¨Ïò®Îã§.
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# train dataÏôÄ test dataÏùò Í∏∏Ïù¥
train.shape, test.shape
```

- train dataÎäî Ïª¨Îüº 785Í∞ú, Î°úÏö∞ 42000Í∞ú
- test dataÎäî Ïª¨Îüº 784Í∞ú, Î°úÏö∞ 28000Í∞ú

```python
train.head()
```

- train dataÏóêÎäî labelÏù¥ Ìè¨Ìï®ÎêòÏñ¥ ÏûàÎã§.

```python
# train dataÏùò labelÏª¨ÎüºÏùÑ Î∂ÑÎ¶¨ÌïúÎã§.
y = train['label']
X = train.drop('label',axis = 1)
```

```python
X.shape, y.shape
```

```python
# test dataÏôÄ train dataÎ•º Î∂ÑÎ¶¨ÌïòÍ∏∞ ÏúÑÌï¥ sklearnÏùò train_test_splitÎ•º importÌïúÎã§.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# dataÍ∞Ä Ïûò ÎÇòÎà†Ï°åÎäîÏßÄ
X_train
```

```python
y_train
```

```python
# ÎûúÎç§ Ìè¨Î†àÏä§Ìä∏Î°ú ÏòàÏ∏°ÌïòÍ∏∞
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_clf = RandomForestClassifier(random_state=42)
rf_clf = rf_clf.fit(X_train, y_train)
rf_prediction = rf_clf.predict(X_test)
print(rf_prediction[0:5])
print(y_test[0:5])
print(accuracy_score(y_test, rf_prediction))
```

```python
# k-NN ÏïåÍ≥†Î¶¨Ï¶òÏúºÎ°ú ÏòàÏ∏°ÌïòÍ∏∞
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train.values.ravel())
```

```python
# ÏòàÏ∏° Ìï¥Î≥¥Í∏∞
prediction = knn_clf.predict(X_test)
```

```python
# Ï†ïÌôïÎèÑ
print(accuracy_score(y_test,prediction))
```

```python
# testÏÖãÏúºÎ°ú Í≤∞Í≥º ÏòàÏ∏°
result=knn_clf.predict(test)
```
